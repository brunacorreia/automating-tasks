# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/1mhXZ3JPAnekXP_4vX7Z_sJj35VWqayaR?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
#importar bibliotecas
#para instalar, abre-se o prompt anaconda e se escreve 'pip install pyautogui' e depois'pip install pyperclip'

import pyautogui # faz automacao moouse e teclado
import time # controla o tempo do programa
import pyperclip # permite copiar e colar com o python

pyautogui.PAUSE = 1
#passo 1: entrar no sistema (Google Drive)
pyautogui.hotkey('ctrl','t')
link = ('https://drive.google.com/drive/u/1/folders/1PtyocK3u2jwxQawBTPVb7MHFgjdcLd_4')
pyautogui.write(link)
pyautogui.press('enter')
#esperar um pouquinho para dar tempo de pegar a localização da pasta
time.sleep(5)

#passo 2: entrar na pasta Aula 1
pyautogui.click(462, 532)

#passo 3: fazer o download da planilha Base de Vendas
pyautogui.click(1659, 230)
pyautogui.click(1456, 738)

time.sleep(5)
pyautogui.click(968, 932)
```

```python
time.sleep(10)
pyautogui.position()
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
#passo 4: calcular os indicadores (faturamento e qtd vendas)

import pandas as pd
tabela = pd.read_excel(r"C:\Users\Acer\Documents\TI\python\hashtag treinamentos\Jupyter\Aula 1 - Vendas Dez.xlsx")
display(tabela)
faturamento = tabela["Valor Final"].sum()
qtde_produtos = tabela["Quantidade"].sum()
```

### Vamos agora enviar um e-mail pelo gmail

```python
#passo 5: entrar no email
pyautogui.hotkey('ctrl', 't')
linkEmail = ('https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox')
pyautogui.write(linkEmail)
pyautogui.press('enter')
time.sleep(5)

#passo 6: criar email
pyautogui.click(87, 223)
time.sleep(5)

#escrever destinatario
pyautogui.write("brunamourac.cursos@gmail.com")
pyautogui.press('tab')
#escrever assunto
pyautogui.press('tab')
assunto = "Relatório de Vendas"
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')

#escrever corpo do email
pyautogui.press('tab')
texto_email = f"""Bom dia, 

Conforme solicitado, segue abaixo as informações de faturamento do dia anterior:
     - Valor total faturado: R$ {faturamento:,.2f}
     - Quantidade de produtos vendidos: {qtde_produtos:,.2f}
     
Atenciosamente,
     
Bruna Moura Correia"""

pyperclip.copy(texto_email)
time.sleep(1)
pyautogui.hotkey('ctrl', 'v')
time.sleep(2)

#passo 7: enviar email

pyautogui.hotkey('ctrl', 'enter')

```
